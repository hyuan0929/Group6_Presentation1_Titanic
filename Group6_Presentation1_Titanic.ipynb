## Challenge 3 - Understanding the Data

The Titanic dataset contains passenger records from the Titanic disaster.  
Key columns include:  
- **Survived**: whether the passenger survived (0 = No, 1 = Yes)  
- **Pclass**: ticket class (1 = first, 2 = second, 3 = third)  
- **Sex**: passenger gender  
- **Age**: passenger age in years  
- **Siblings/Spouses Aboard**: number of siblings or spouses on board  
- **Parents/Children Aboard**: number of parents or children on board  
- **Fare**: price of the passenger’s ticket  

This dataset is useful for studying survival patterns based on class, gender, age, and fare.


```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3

class DataExplorer:
    """
    A helper class for data exploration and visualization.
    """

    def __init__(self, dataframe: pd.DataFrame):
        self.df = dataframe

    # === Descriptive Statistics ===
    def calculate_mean(self, column: str) -> float:
        return self.df[column].mean()

    def calculate_median(self, column: str) -> float:
        return self.df[column].median()

    def calculate_mode(self, column: str):
        return self.df[column].mode().tolist()

    def calculate_variance(self, column: str) -> float:
        return self.df[column].var()

    def calculate_std(self, column: str) -> float:
        return self.df[column].std()

    def calculate_quartiles(self, column: str):
        return self.df[column].quantile([0.25, 0.5, 0.75]).to_dict()

    # === Visualization Methods ===
    def scatter_plot(self, x: str, y: str):
        plt.figure()
        self.df.plot.scatter(x=x, y=y, alpha=0.5)
        plt.title(f"Scatter Plot: {x} vs {y}")
        plt.show()

    def histogram(self, column: str, bins: int = 10):
        plt.figure()
        self.df[column].plot.hist(bins=bins, alpha=0.7)
        plt.title(f"Histogram of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()

    def box_plot(self, column: str, by: str = None):
        plt.figure()
        self.df.boxplot(column=column, by=by)
        plt.title(f"Boxplot of {column} by {by}" if by else f"Boxplot of {column}")
        plt.suptitle("")  # remove extra title
        plt.show()

    def venn_diagram(self, set1, set2, set3=None, labels=("Set1", "Set2", "Set3")):
        plt.figure()
        if set3:
            venn3([set1, set2, set3], set_labels=labels)
        else:
            venn2([set1, set2], set_labels=labels)
        plt.title("Venn Diagram")
        plt.show()
```

### Test the DataExplorer class


```python
# Initialize DataExplorer
explorer = DataExplorer(df)

# Example: calculate the mean of Age
print("Mean Age:", explorer.calculate_mean("Age"))

# Example: draw a scatter plot of Age vs Fare
explorer.scatter_plot("Age", "Fare")
```

## Challenge 5 - Central Tendency and 50-word Summary

We calculate the **mean**, **median**, and **mode** of key variables such as Age and Fare.  
These measures of central tendency help us understand the general distribution of passengers’ demographics and ticket prices.  
The following results provide insight into the overall population structure on the Titanic.


```python
# Calculate mean, median, and mode for Age and Fare
print("Age - Mean:", explorer.calculate_mean("Age"))
print("Age - Median:", explorer.calculate_median("Age"))
print("Age - Mode:", explorer.calculate_mode("Age"))

print("\nFare - Mean:", explorer.calculate_mean("Fare"))
print("Fare - Median:", explorer.calculate_median("Fare"))
print("Fare - Mode:", explorer.calculate_mode("Fare")[:5])  # only show first 5 modes
```

### 50-word Summary

The Titanic dataset shows that the average passenger age was about 29 years, with most fares concentrated below 40 units.  
The median values highlight how age and ticket price were skewed by outliers.  
Overall, the measures of central tendency reveal meaningful insights into passenger demographics and economic status.


## Challenge 6 - Measures of Dispersion


```python
# Calculate variance, standard deviation, and quartiles for Age
print("Age - Variance:", explorer.calculate_variance("Age"))
print("Age - Standard Deviation:", explorer.calculate_std("Age"))
print("Age - Quartiles:", explorer.calculate_quartiles("Age"))

# Calculate variance, standard deviation, and quartiles for Fare
print("\nFare - Variance:", explorer.calculate_variance("Fare"))
print("Fare - Standard Deviation:", explorer.calculate_std("Fare"))
print("Fare - Quartiles:", explorer.calculate_quartiles("Fare"))
```

### Summary
The results show that Fare has much higher variance and standard deviation compared to Age,  
indicating that ticket prices varied significantly among passengers.  
Quartiles further highlight the skewness of Fare, where most values are concentrated at the lower range with a few extreme outliers.


## Challenge 7 - Visualization and Numerical Summary
